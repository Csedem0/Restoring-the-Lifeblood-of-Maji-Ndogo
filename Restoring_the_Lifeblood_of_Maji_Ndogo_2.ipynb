### Employee table. It has info on all of our workers, but note that the email addresses have not been added. We will have to send them reports and figures, so let's update it. Luckily the emails for our department are easy: first_name.last_name@ndogowater.gov.

### Method 1

```python
%%sql


SELECT * FROM md_water_services.employee LIMIT 5;
```

```python
%%sql

SELECT
    employee_name,
    LENGTH(employee_name) AS string_length,
    POSITION(' ' IN employee_name) AS post_first_name,
    LEFT(employee_name, POSITION(' ' IN employee_name) -1) AS first_name,
    LENGTH(LEFT(employee_name, POSITION(' ' IN employee_name) -1)) AS first_name_len
FROM 
    md_water_services.employee LIMIT 10;
```

```python
%%sql

SELECT
    employee_name,
    LENGTH(employee_name) AS string_length,

    SUBSTRING(
        employee_name
        FROM POSITION(' ' IN employee_name) + 1
    ) AS last_name,

    LENGTH(
        SUBSTRING(
            employee_name
            FROM POSITION(' ' IN employee_name) + 1
        )
    ) AS last_name_len

FROM md_water_services.employee
LIMIT 10;
```

```python
%%sql

SELECT
    employee_name,
    LOWER(CONCAT(LEFT(employee_name, POSITION(' ' IN employee_name) -1), '.',  SUBSTRING(employee_name FROM POSITION(' ' IN employee_name) + 1), '@ndogowater.gov.')) AS email
FROM 
    md_water_services.employee
LIMIT 10;
    
```

```python
%%sql


# Query that counts the number of records per town

SELECT
    COUNT(location_id) AS records_per_town,
    town_name
FROM 
    location
GROUP BY 
    town_name
ORDER BY records_per_town DESC
LIMIT 8;
```

```python
%%sql


# Counting the records per province.

SELECT
    COUNT(location_id) AS records_per_province,
    province_name
FROM 
    location
GROUP BY 
    province_name
ORDER BY records_per_province DESC;
```

```python
%%sql


# Counting the records per town and per province.

SELECT
    province_name,
    town_name,
    COUNT(location_id) AS records_per_town
FROM 
    location
GROUP BY 
    province_name, town_name
ORDER BY 
    province_name, records_per_town DESC
LIMIT
    10
```

```python
%%sql


# Counting the records per location.

SELECT
    COUNT(location_id) AS num_sources,
    location_type
FROM 
    location
GROUP BY location_type;
```

#### Total number of people we surveyed

```python
%%sql

SELECT
    SUM(number_of_people_served) AS Total_persons_surveyed
FROM 
    water_source
```

#### Total number of well, taps and rivers

```python
%%sql

SELECT
    type_of_water_source,
    COUNT(number_of_people_served) AS number_of_sources
FROM 
    water_source
GROUP BY type_of_water_source
```

#### Average number of people that are served by each water source

```python
%%sql

SELECT
    type_of_water_source,
    ROUND(AVG(number_of_people_served), 0) AS ave_people_per_source
FROM 
    water_source
GROUP BY type_of_water_source
```

#### The total number of people served by each type of water source in total.

```python
%%sql

SELECT
    type_of_water_source,
    SUM(number_of_people_served) AS population_served
FROM 
    water_source
GROUP BY type_of_water_source
ORDER BY population_served DESC
```

#### The total percentage of people served by each type of water source in total.

```python
%%sql

SELECT
    type_of_water_source,
    ROUND((SUM(number_of_people_served)/27000000)*100, 0) AS percentage_people_per_source
FROM 
    water_source
GROUP BY type_of_water_source
ORDER BY percentage_people_per_source DESC
```

## Start of a solution
### At some point, we will have to fix or improve all of the infrastructure, so we should start thinking about how we can make a data-driven decision
### My simple approach is to fix the things that affect most people first.

```python
%%sql

SELECT
    type_of_water_source,
    SUM(number_of_people_served) AS population_served,
    RANK() OVER( ORDER BY (SUM(number_of_people_served ) )DESC) AS rank_by_population
FROM 
    water_source 
WHERE type_of_water_source != 'tap_in_home'
GROUP BY type_of_water_source
```

```python
%%sql

SELECT
    *,
    RANK() OVER( ORDER BY type_of_water_source, number_of_people_served) AS priority_rank
FROM
water_source
WHERE type_of_water_source != 'tap_in_home'
ORDER BY priority_rank 
LIMIT 6
```

#### 2. What is the average total queue time for water?

```python
%%sql

SELECT
    ROUND(AVG(NULLIF(time_in_queue, 0))) AS average_total_queue_time
FROM
visits
```

```python
%%sql

SELECT
    DAYNAME(time_of_record) AS day_of_week,
    ROUND(AVG(NULLIF(time_in_queue, 0))) AS average_total_queue_time
FROM
visits
GROUP BY day_of_week
```

```python
%%sql

SELECT
TIME_FORMAT(TIME(time_of_record), '%H:00') AS hour_of_day,
ROUND(AVG(NULLIF(time_in_queue, 0))) AS average_total_queue_time
FROM
visits
GROUP BY hour_of_day
```
